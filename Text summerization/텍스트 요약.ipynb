```python
# 한글만 추출하기
def hangul_only(df : pd.DataFrame, convert_column : str) -> pd.DataFrame:
    df['sentence'] = df[convert_column].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣0-9 ]","")
    df['sentence'] = df['sentence'].replace('^ +', '')
    df['sentence'].replace('', np.nan, inplace = True)
    df = df.dropna(how = 'any')
    return df

melon_hangul = hangul_only(melon_data, 'lyric')
```

```python
# 한글 추출 설정
melon_hangul
```

```python
# model 불러오기
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration
tokenizer = PreTrainedTokenizerFast.from_pretrained('ainize/kobart-news')
model = BartForConditionalGeneration.from_pretrained('ainize/kobart-news')
```

```python
lyrics = melon_hangul['sentence']
lyrics
```

```python
bin_list = []
for i in range(len(lyrics)):
  input_ids = tokenizer.encode(lyrics[i], return_tensors = "pt")
  summary_lyrics = model.generate(
      input_ids = input_ids,
      bos_token_id = model.config.bos_token_id,
      eos_token_id = model.config.eos_token_id,
      length_penalty = 1.0,
      max_length = 142,
      min_length = 56,
      num_beams = 4,
  )
  decode_lyrics = tokenizer.decode(summary_lyrics[0], skip_special_tokens=True)
  bin_list.append(decode_lyrics)
  print(i)
```

```python
melon_hangul['lyric_summary'] = bin_list
melon_hangul
```
